```python
import pandas as pd
data = pd.read_csv("House Price India.csv")

data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

correlation_matrix = data.corr()
plt.figure(figsize=(20,10))
sb.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x = data.drop(columns=["Price"])
y = data["Price"]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2 , random_state=42)
print(x_train, x_test, y_train, y_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100 , random_state=42)
rf.fit(x_train, y_train)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor(n_estimators=100 , random_state=42)
gb.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
#Linear Regression Evaluation
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("MSE of Linear Regression : " ,mse)
print("R2 Score of Linear Regression : ",r2*100,"%")
```

```python
#Random Forest Evaluation Regressor
rf_pred = model.predict(x_test)
rf_mse = mean_squared_error(y_test,rf_pred)
rf_r2 = r2_score(y_test,rf_pred)
print("MSE of Random Forest Regression : ",rf_mse)
print("R2 Score of Random Forest Regression : ",rf_r2*100,"%") 
```

```python
#Gradient Booster Regressor
gb_pred = model.predict(x_test)
gb_mse = mean_squared_error(y_test,gb_pred)
gb_r2 = r2_score(y_test, gb_pred)
print("MSE of Gradient Booster Regression : " ,gb_mse)
print("R2 of Gradient Booster Regression : " , gb_r2*100,"%")
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 200],  # Reduce number of estimators
    'max_depth': [10, 20],  # Avoid using None
    'min_samples_split': [2, 5]  # Reduce options
}
grid_search = GridSearchCV(RandomForestRegressor(), param_grid, cv=3, n_jobs=-1, verbose=2)
grid_search.fit(x_train, y_train)
print(grid_search.best_params_)
```
